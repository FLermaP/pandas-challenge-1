## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df['line_subtotal']=df['unit_price']*df['qty']
df[['unit_price','qty','line_subtotal']].head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
def shipping_price(weight):
    if weight>50 : return weight*7
    return weight*10
df['total_weight']=df['unit_weight']*df['qty']
df['shipping_price']=df['total_weight'].apply(shipping_price)
df[['unit_price','unit_weight','qty','total_weight','shipping_price']].head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
import numpy as np
df['line_price']=np.round((df['line_subtotal']+df['shipping_price'])*1.0925,decimals=2)
df[['line_subtotal','shipping_price','line_price']].head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df['line_cost']=(df['unit_cost']*df['qty'])+df['shipping_price']
df.head()
```

```python
# Create a column for the profit of
# each line using line cost and line price
df['line_profit']=df['line_price']-df['line_cost']
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.

## Create a new df with just client id
unique_id=pd.DataFrame({'client_id':list(df['client_id'].unique())})
```

```python
## Add a column to store the total qty purchased
unique_id.insert(1,'total_qty',value=0)
```

```python
## Calculate (sum) the quantity by client
i=0
for client in unique_id['client_id']:
    unique_id['total_qty'][i]=df.loc[df['client_id']==unique_id['client_id'][i],'qty'].sum()
    i+=1
```

```python
## Create a new DF with just the top 5 clients by qty 
top_5_by_qty=unique_id.sort_values('total_qty', ascending=False).head(5).copy()
```

```python
## Reset the index to be able to refer to by index
top_5_by_qty=top_5_by_qty.reset_index(drop=True)
```

```python
## Add a column to store the total spent
top_5_by_qty.insert(2,'total_spent',value=0.0)
```

```python
## Calculate (sum) the amount spent by client
i=0
for client in top_5_by_qty['client_id']:
    top_5_by_qty['total_spent'][i]=df.loc[df['client_id']==top_5_by_qty['client_id'][i],'line_price'].sum()
    i+=1
```

```python
i=0
for client in top_5_by_qty['client_id']:
    print(f"{top_5_by_qty['client_id'][i]} spent $ {top_5_by_qty['total_spent'][i]:,.2f}")
    i+=1
# unique_id
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
def new_totals(destination, source):
    i=0
    for client in top_5_by_qty['client_id']:
        top_5_by_qty[destination][i]=df.loc[df['client_id']==top_5_by_qty['client_id'][i],source].sum()
        i+=1
```

```python
top_5_by_qty.insert(2,'total_shipping',value=0.0)
```

```python
new_totals('total_shipping','shipping_price')
```

```python
top_5_by_qty.insert(4,'total_profit',value=0.0)
```

```python
new_totals('total_profit','line_profit')
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
top_5_by_qty=top_5_by_qty.rename(columns={'client_id':'Client ID', 'total_qty':'Purchased Quantity',\
  'total_shipping':'Shipping Cost (MM)','total_spent':'Revenue from Client (MM)', 'total_profit':'Profit from Client (MM)'})
```

```python
top_5_by_qty.sort_values('Profit from Client (MM)', inplace=True)
```

```python
top_5_by_qty['Purchased Quantity']=top_5_by_qty['Purchased Quantity'].map('{:,}'.format)
top_5_by_qty['Shipping Cost (MM)']=top_5_by_qty['Shipping Cost (MM)']/1000000
top_5_by_qty['Shipping Cost (MM)']=top_5_by_qty['Shipping Cost (MM)'].map('$ {:.2f}'.format)
top_5_by_qty['Revenue from Client (MM)']=top_5_by_qty['Revenue from Client (MM)']/1000000
top_5_by_qty['Revenue from Client (MM)']=top_5_by_qty['Revenue from Client (MM)'].map('$ {:.2f}'.format)
top_5_by_qty['Profit from Client (MM)']=top_5_by_qty['Profit from Client (MM)']/1000000
top_5_by_qty['Profit from Client (MM)']=top_5_by_qty['Profit from Client (MM)'].map('$ {:.2f}'.format)
```
